```python
# Student marks analysis program
# Using loops and conditionals to calculate grades and stats

student_name = "Ayesha Sharma"
subjects = ["Math", "Physics", "CS", "English", "Chemistry"]
marks = [88, 72, 95, 60, 78]
```

```python
for i in range(len(subjects)):   # loop through all subjects
    subject = subjects[i]
    mark = marks[i]

    # assigning grade based on marks
    if mark >= 90:
        grade = "A+"
    elif mark >= 80:
        grade = "A"
    elif mark >= 70:
        grade = "B"
    elif mark >= 60:
        grade = "C"
    else:
        grade = "F"

    print(subject, ":", mark, "-", grade)  # display result
```

```python
# calculating total and average marks
total = sum(marks)
average = total / len(marks)

print("Total Marks:", total)
print("Average Marks:", round(average, 2))
```

```python
# finding highest marks
max_marks = max(marks)
max_index = marks.index(max_marks)

# finding lowest marks
min_marks = min(marks)
min_index = marks.index(min_marks)
```

```python
new_count = 0   # count of new subjects added

while True:
    new_subject = input("Enter subject name (or type 'done' to stop): ")

    if new_subject.lower() == "done":   # stop condition
        break

    new_marks = input("Enter marks (0-100): ")

    # checking if input is numeric
    if not new_marks.isdigit():
        print("Invalid input!")
        continue

    new_marks = int(new_marks)

    # checking valid range
    if new_marks < 0 or new_marks > 100:
        print("Marks should be between 0 and 100")
        continue

    # adding valid data to list
    subjects.append(new_subject)
    marks.append(new_marks)
    new_count += 1
```

```python
# printing final results
print("New subjects added:", new_count)

new_total = sum(marks)
new_average = new_total / len(marks)

print("Updated Average:", round(new_average, 2))
```
