```python
# =========================================
# Task 1 — Data Parsing & Profile Cleaning
# =========================================

raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

# Loop through each student
for student in raw_students:

    # name clean (remove spaces + title case)
    name_clean = student["name"].strip().title()

    # Convert roll number to integer
    roll_no = int(student["roll"])

    # Convert marks string to list of integers
    marks_list_final = student["marks_str"].split(", ")
    marks = []

    for m in marks_list_final:
        marks.append(int(m))

    # Check if name is valid (only alphabets)
    valid = True
    for word in name_clean.split():
        if not word.isalpha():
            valid = False

    # Print validation result
    if valid:
        print("✓ Valid name")
    else:
        print("✗ Invalid name")

    # Print formatted profile
    print("="*40)
    print(f"Student : {name_clean}")
    print(f"Roll No : {roll_no}")
    print(f"Marks   : {marks}")
    print("="*40)


# Find student with roll 103
for student in raw_students:
    if int(student["roll"]) == 103:
        name = student["name"].strip().title()
        print(name.upper())
        print(name.lower())
```

```python
# =========================================
# Task 2 — Marks Analysis
# =========================================

student_name = "Ayesha Sharma"
subjects_list = ["Math", "Physics", "CS", "English", "Chemistry"]
marks = [88, 72, 95, 60, 78]

# 1. Print subjects with grades
for i in range(len(subjects_list)):
    subject = subjects_list[i]
    mark = marks[i]

    # deciding grade based on marks
    if mark >= 90:
        grade = "A+"
    elif mark >= 80:
        grade = "A"
    elif mark >= 70:
        grade = "B"
    elif mark >= 60:
        grade = "C"
    else:
        grade = "F"

    print(f"{subject} -> {mark} | Grade: {grade}")


# 2. Total & Average
total = sum(marks)
average = round(total / len(marks), 2)

print("Total Marks:", total)
print("Average:", average)


# Highest & Lowest
max_mark = max(marks)
min_mark = min(marks)

max_index = marks.index(max_mark)
min_index = marks.index(min_mark)

print("Highest:", subjects_list[max_index], "-", max_mark)
print("Lowest:", subjects_list[min_index], "-", min_mark)


# 3. While loop (add new subjects)
new_count = 0

while True:
    sub = input("Enter subject name (or type 'done' to stop): ")

    if sub.lower() == "done":
        break

    mark_input = input("Enter marks (0-100): ")

    # checking valid number
    if not mark_input.isdigit():
        print("Please enter valid number!")
        continue

    mark_input = int(mark_input)

    if mark_input < 0 or mark_input > 100:
        print("Marks should be between 0 and 100!")
        continue

    # adding new subject
    subjects_list.append(sub)
    marks.append(mark_input)
    new_count += 1


# Final output
print("Total new subjects:", new_count)

new_avg = round(sum(marks) / len(marks), 2)
print("Updated Average:", new_avg)
```

```python
# =========================================
# Task 3 — Class Performance Summary
# =========================================

class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

passed_student = 0
failed_student = 0
total_avg_sum = 0

topper_name = ""
topper_avg = 0

print("Name              | Average | Status")
print("-"*40)

# loop through each student
for name, marks in class_data:

    avg = round(sum(marks) / len(marks), 2)
    total_avg_sum += avg

    # pass or fail
    if avg >= 60:
        status = "Pass"
        passed_student += 1
    else:
        status = "Fail"
        failed_student += 1

    # check topper
    if avg > topper_avg:
        topper_avg = avg
        topper_name = name

    # print row
    print(f"{name:<18} | {avg:<7} | {status}")

# final outputs
print("-"*40)
print("Passed:", passed_student)
print("Failed:", failed_student)

print("Topper:", topper_name, "-", topper_avg)

class_avg = round(total_avg_sum / len(class_data), 2)
print("Class Average:", class_avg)
```

```python
# =========================================
# Task 4 — String Manipulation Utility
# =========================================

essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

# 1. remove extra spaces
final_text = essay.strip()
print("final_text:")
print(final_text)

# 2. convert to title case
title_version = final_text.title()
print("\nTitle Case:")
print(title_version)

# 3. count word 'python'
count_python = final_text.count("python")
print("\nCount of 'python':", count_python)

# 4. replace 'python' with 'Python 🐍'
replaced_text = final_text.replace("python", "Python 🐍")
print("\nReplaced Text:")
print(replaced_text)

# 5. split into sentences
sentences = final_text.split(". ")
print("\nSentence List:")
print(sentences)

# 6. print sentences with numbering
print("\nNumbered Sentences:")
for i in range(len(sentences)):
    sentence = sentences[i]

    # add full stop if missing
    if not sentence.endswith("."):
        sentence = sentence + "."

    print(f"{i+1}. {sentence}")
```
